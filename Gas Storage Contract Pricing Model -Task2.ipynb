Q4. "Can you calculate the profit or value of a gas storage contract based on buying, storing, and selling gas?"

```python
# Function to calculate storage contract value

def price_gas_contract(injection_dates,
                       withdrawal_dates,
                       injection_prices,
                       withdrawal_prices,
                       injection_rate,
                       withdrawal_rate,
                       max_storage_volume,
                       storage_cost_per_unit):

    # Total profit initialization
    total_profit = 0

    # Current storage volume
    stored_volume = 0

    # Loop through injection transactions
    for i in range(len(injection_dates)):

# Gas volume injected -How much gas can still fit in storage?
        injected_volume = min(
            injection_rate,
            max_storage_volume - stored_volume
        )

    # Update storage volume -Adds injected gas into storage.
        stored_volume += injected_volume

        # Purchase cost -Calculates money spent buying gas.
        purchase_cost = injected_volume * injection_prices[i]

    # Add negative cash flow -Buying gas is an expense so Profit decreases.
        total_profit -= purchase_cost

        # Storage cost
        total_profit -= injected_volume * storage_cost_per_unit

    # Loop through withdrawal transactions
    # Gas is removed from storage and sold.
    for i in range(len(withdrawal_dates)):

# Gas volume withdrawn -Cannot withdraw more gas than available.
        withdrawn_volume = min(
            withdrawal_rate,
            stored_volume
        )

        # Update storage volume -Remove sold gas from storage.
        stored_volume -= withdrawn_volume

        # Revenue from selling gas -Calculates revenue earned.
        sale_revenue = withdrawn_volume * withdrawal_prices[i]

        # Add positive cash flow -Selling gas generates income.
        total_profit += sale_revenue

    # Return final contract value
    return total_profit
```

Q5. "What contract details should be provided to calculate the value of a gas storage contract?"

```python
# Sample inputs

injection_dates = ["2024-01-01", "2024-02-01"]

withdrawal_dates = ["2024-06-01", "2024-07-01"]

injection_prices = [20, 22]

withdrawal_prices = [28, 30]

injection_rate = 1000

withdrawal_rate = 1000

max_storage_volume = 2000

storage_cost_per_unit = 1
```

Q6. "What is the final profit or value generated from the gas storage contract?"

```python
# Calculate and display the final value of the gas storage contract

contract_value = price_gas_contract(
    injection_dates,
    withdrawal_dates,
    injection_prices,
    withdrawal_prices,
    injection_rate,
    withdrawal_rate,
    max_storage_volume,
    storage_cost_per_unit
)
# Display result

print("Contract Value:", contract_value)
```
